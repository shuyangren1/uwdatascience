```python
# import the nessessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# loading in data from assignment URL
url = "https://library.startlearninglabs.uw.edu/DATASCI410/Datasets/JitteredHeadCount.csv"
casino = pd.read_csv(url)
casino.head(10)
```

```python
sns.catplot(x="DayOfWeek", y="HeadCount", hue="GameCode", kind="bar", data=casino);
plt.title('Head Count by Day of the Week');
plt.ylabel('Number of Occupied Tables')
plt.xlabel('Day of the Week')
```

Observing the data, we see that games S6, CR, and TP (Big Six, Craps, and Texas Hol'em Poker) are among the most popular. Games are overall more popular on weekends (Fri, Sat, Sun) with Saturday being the busiest.
